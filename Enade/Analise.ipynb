```python
##Percepção e tempo gasto dos alunos da FATEC-SJC do curso de Análise e Desenvolvimento de Sistemas para fazer o ENADE 2017

##Desenvolvido por Wesley Dias

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('2017/3.DADOS/MICRODADOS_ENADE_2017.txt', sep=';', low_memory=False, encoding='ANSI', usecols=['CO_IES', 'CO_GRUPO','CO_RS_I1', 'CO_RS_I2', 'CO_RS_I9'])

def Contar_Respostas(questao):
    Cont_Res = [0, 0, 0, 0, 0]
    for resposta in df.query('CO_IES == 15581 and CO_GRUPO == 72')[questao]:
        if resposta == 'A': Cont_Res[0] += 1
        elif resposta == 'B': Cont_Res[1] += 1
        elif resposta == 'C': Cont_Res[2] += 1
        elif resposta == 'D': Cont_Res[3] += 1
        elif resposta == 'E': Cont_Res[4] += 1
    return Cont_Res

def Criar_Grafico(x, y, titulo='Trocar'):
    plt.xlabel('Dificuldade')
    plt.ylabel('Número de alunos')
    plt.title(titulo)
    plt.bar(x, y, color=['#3CB371', '#98FB98', '#F0E68C', '#FA8072', '#B22222'])

Resp_Dif = ['Muito Fácil', 'Fácil', 'Médio', 'Difícil', 'Muito Difícil']
Resp_Temp = ['< 1', '1 e 2', '2 e 3', '3 e 4', '4 e Não Term. ']
q1 = Contar_Respostas('CO_RS_I1')
q2 = Contar_Respostas('CO_RS_I2')
q3 = Contar_Respostas('CO_RS_I9')
```

```python
Criar_Grafico(Resp_Dif, q1, titulo='FATEC-SJC (ADS) Dificuldade na parte de Formação Geral (2017)')
print(q1)
```

```python
Criar_Grafico(Resp_Dif, q2, titulo='FATEC-SJC (ADS) Dificuldade na parte do Componente Específico (2017)')
print(q2)
```

```python
Criar_Grafico(Resp_Temp, q3, titulo='FATEC-SJC (ADS) Tempo gasto para fazer a prova (2017)')
plt.xlabel('Tempo gasto para fazer a prova (Horas)')
print(q3)
```
